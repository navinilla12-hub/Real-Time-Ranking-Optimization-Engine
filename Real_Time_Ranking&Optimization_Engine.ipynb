```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "item_id": range(1, 11),
    "score": np.random.rand(10),
    "bid": np.random.rand(10)
})
```

```python
df["ranking_score"] = df["score"] * 0.7 + df["bid"] * 0.3
```

```python
df = df.sort_values("ranking_score", ascending=False)

df
```
